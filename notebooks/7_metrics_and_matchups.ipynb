```python
conds = [
    games.h_score > games.a_score,
    games.h_score < games.a_score,
    (games.rnd == 'Final') & (games.h_score == games.a_score) & (games.h_pens > games.a_pens),
    (games.rnd == 'Final') & (games.h_score == games.a_score) & (games.h_pens < games.a_pens)
]

values = [games.h_team, games.a_team, games.h_team, games.a_team]

games['game_winner'] = np.select(conds, values, 'tie')
```

```python
# Part of what we are trying to find is away goal luck, which means we need to figure out how to combine home and away matches.
# i think i'll need a col with the teams invovled
def teamorder(x):
    if (x.h_coef < 1) or (x.a_coef < 1) or (x.h_coef == x.a_coef):
        if x.h_team < x.a_team:
            return x.h_team + "," + x.a_team
        else:
            return x.a_team + "," + x.h_team
    if x.h_coef > x.a_coef:
        return x.h_team + "," + x.a_team
    if x.h_coef < x.a_coef:
        return x.a_team + "," + x.h_team

games['teams'] = games.apply(lambda x: teamorder(x), axis=1)
```

```python
matchups = pd.DataFrame()
for k, v in matchups_dict.items():
    game1 = v.iloc[[0]]
    game2 = v.iloc[[1]]
    m = game1.merge(game2, on=['season', 'rnd', 'teams'], suffixes=['_g1', '_g2'])
    matchups = matchups.append(m)
```

```python
# I want to add the winner and total goals for each team
# team1 is the home team of the first game
matchups['team1_total'] = matchups.h_score_g1 + matchups.a_score_g2
matchups['team2_total'] = matchups.a_score_g1 + matchups.h_score_g2
```

```python
# add total goals in a matchup
matchups['total_goals'] = matchups.team1_total + matchups.team2_total
```

```python
# add the win condition
conds = [
    (matchups['team1_total'] > matchups['team2_total']) | (matchups['team1_total'] < matchups['team2_total']),
    (matchups['team1_total'] == matchups['team2_total']) & (matchups.h_pens_g2 > 0),
    (matchups['team1_total'] == matchups['team2_total']) & (matchups.a_score_g1 - matchups.a_score_g2 != 0)
]

vals = [
    'total',
    'pens',
    'away'
]

matchups['win_cond'] = np.select(conds, vals)
```

```python
matchups
```

```python
# add the winner and loser
conds = [
    (matchups['team1_total'] > matchups['team2_total']), 
    (matchups['team1_total'] < matchups['team2_total']),
    ((matchups['team1_total'] == matchups['team2_total']) & (matchups.a_score_g1 < matchups.a_score_g2)),
    ((matchups['team1_total'] == matchups['team2_total']) & (matchups.a_score_g1 > matchups.a_score_g2)),
    ((matchups['team1_total'] == matchups['team2_total']) & (matchups.h_pens_g2 < matchups.a_pens_g2)),
    ((matchups['team1_total'] == matchups['team2_total']) & (matchups.h_pens_g2 > matchups.a_pens_g2))

]

vals = [
    matchups.h_team_g1,
    matchups.a_team_g1,
    matchups.h_team_g1,
    matchups.a_team_g1,
    matchups.h_team_g1,
    matchups.a_team_g1
]

vals_loser = [
    matchups.a_team_g1,
    matchups.h_team_g1,
    matchups.a_team_g1,
    matchups.h_team_g1,
    matchups.a_team_g1,
    matchups.h_team_g1
]

matchups['winner'] = np.select(conds, vals)
matchups['loser'] = np.select(conds, vals_loser)
```
